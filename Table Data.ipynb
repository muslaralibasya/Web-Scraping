```python
#install library and download webdriver
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
import time
from datetime import datetime as dt
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',options=options)
```

```python
url = 'https://hallo.palembang.go.id/covid19/'
driver.get(url)
time.sleep(10)
driver.find_element_by_xpath('//*[@id="examplez_length"]/label/select/option[4]').click()
time.sleep(10)
content = driver.page_source
time.sleep(10)
soup = BeautifulSoup(content, "html.parser")
div = soup.find('table', class_='table table-bordered table-hover dt-responsive ftsize dataTable no-footer dtr-inline collapsed')
links = div.find_all('td')
```

```python
#find time and date
now = dt.now().strftime('%Y-%m-%d')
```

```python
#create list
scrape_date = []
date_update = []
total_odp = []
total_pdp = []
total_positif = []
positif_sembuh = []
positif_isolasi = []
positif_dirawat = []
positif_meninggal = []
source_link = []
kecamatan = []
```

```python
#Append data to list
for (j, link) in enumerate(links):
  if j in range(1, len(links), 13):
    kecamatan.append(link.get_text())
  if j in range(3, len(links), 13):
    total_odp.append(link.get_text())
  if j in range(6, len(links), 13):
    total_pdp.append(link.get_text())
  if j in range(9, len(links), 13):
    total_positif.append(link.get_text())
  if j in range(10, len(links), 13):
    positif_dirawat.append(link.get_text())
  if j in range(11, len(links), 13):
    positif_sembuh.append(link.get_text())
  if j in range(12, len(links), 13):
    positif_meninggal.append(link.get_text())
  if j in range(12, len(links), 13):
    positif_isolasi.append('')
  if j in range(1, len(links), 13):
    scrape_date.append(now)
  if j in range(1, len(links), 13):
    date_update.append('')
  if j in range(1, len(links), 13):
    source_link.append(url)
```

```python
#create table
df = pd.DataFrame()
df['scrape_date'] = scrape_date
df['date_update'] = date_update
df['kecamatan'] = kecamatan
df['total_odp'] = total_odp
df['total_pdp'] = total_pdp
df['total_positif'] = total_positif
df['positif_sembuh'] = positif_sembuh
df['positif_dirawat'] = positif_dirawat
df['positif_isolasi'] = positif_isolasi
df['positif_meninggal'] = positif_meninggal
df['source_link'] = source_link
df.insert(2,column='provinsi', value='Sumatera Selatan')
df.insert(3,column='kabkot', value='Kota Palembang')
df.insert(5,column='kelurahan', value='')
df.insert(14,column='types', value='kecamatan')
df.insert(15,column='user_pic', value='Muslar')
```
